```python
from turtle import *

def write_text(x, y, size, c , text):
    penup()
    goto(x, y)
    pendown()
    color(c)
    style = ('b tabassom', size, 'bold italic')
    write(text, font=style, align='center')

def draw_rectangle1(width, height, color, x, y):
  penup()
  goto(x,y)
  fillcolor(color)
  begin_fill()
  pendown()
  
  for a in range(2):
    forward(height)
    right(90)
    forward(width)
    right(90)
  
  end_fill()

  hideturtle()

"""
def draw_rectangle2(width, height, color, x, y):
  penup()
  goto(x,y)
  pendown()
  
  for d in range(2):
    forward(height)
    right(90)
    forward(width)
    right(90)

  hideturtle()

def draw_square(size, color, x, y):
  penup()
  goto(x,y)
  pendown()
  
  for x in range(4):
    forward(size)
    right(90)

  hideturtle()
"""

def draw_circle(radius, color, x, y ):
  penup()
  goto(x,y)
  pendown()
  fillcolor(color)
  begin_fill()
  circle(radius)
  end_fill()
  hideturtle()

draw_rectangle1(200, 560, "yellow", -300, 100)

draw_rectangle1(70, 70, "blue", -280, 80)

draw_rectangle1(70, 70, "blue", -190, 80)

draw_rectangle1(70, 70, "blue", -100, 80)

draw_rectangle1(70, 70, "blue", -10, 80)

draw_rectangle1(70, 70, "blue", 80, 80)

draw_rectangle1(120, 70, "blue", 180, 80)

#draw_rectangle2(200, 400, "blue")

draw_circle(50, "black", -200, -150)

draw_circle(50, "black", 120, -150)

draw_circle(25, "orange", -200, -125)

draw_circle(25, "orange", 120, -125  )

write_text(-30, -80,  30, "pink" , "Asal's buss")

turtle.done()
```
